## Write a program to calc the highest prob of a word (w2) occuring after another word (w1)


```python
from collections import defaultdict

def highest_prob(text):
  words = text.split()

  bigram_counts = defaultdict(lambda: defaultdict(int))
  first_word_counts = defaultdict(int)

  for i in range(len(words)-1):
    first_word = words[i]
    second_word = words[i+1]

    bigram_counts[first_word][second_word] += 1
    first_word_counts[first_word] += 1

  result = {}

  for first_word in bigram_counts:
    max_prob = 0
    best_second_word = None

    for second_word in bigram_counts[first_word]:
      prob = bigram_counts[first_word][second_word] / first_word_counts[first_word]

      if prob > max_prob:
        max_prob = prob
        best_word = second_word

    result[first_word] = (best_word, max_prob)

  return result
```

```python
num = highest_prob(text)

for w1, (w2, prob) in num.items():
  print(f"Word 1: {w1}, Word 2: {w2}, Probability: {prob}")
```

```python
# 3 - a

import nltk
from nltk.collocations import BigramCollocationFinder, TrigramCollocationFinder
from nltk.metrics import BigramAssocMeasures, TrigramAssocMeasures
from nltk.corpus import stopwords

import string

nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
text = """
  Natural language processing is a fascinating area of Artificial Intelligence. It deals with how computers understand and generate human language. Word collocations are pairs of words that appear together more often than by chance. For example, 'artificial intelligence', 'machine learning' and 'deep learning'
"""

tokens = nltk.word_tokenize(text.lower())

stop_words = set(stopwords.words('english'))
filtered_tokens = [t for t in tokens if t not in stop_words and t not in string.punctuation]

print("Top Bigram Collocations")

bigram_finder = BigramCollocationFinder.from_words(filtered_tokens)
bigram_finder.apply_freq_filter(1)


scored_bigrams = bigram_finder.score_ngrams(BigramAssocMeasures.pmi)

for (w1, w2), score in scored_bigrams[:10]:
  print(f"{w1} {w2}: {score}")

```

```python
def is_operator(token):
    return token in ['+', '-', '*', '/', '^', '=', '(', ')']

def is_operand(token):
    # Accepts numbers (with decimals) and alphabetic variables
    return token.replace('.', '', 1).isdigit() or token.isalpha()

def tokenize(text):
    tokens = []
    current = ''

    for ch in text:
        if ch in '+-*/^=()':
            if current:
                tokens.append(current.strip())
                current = ''
            tokens.append(ch)
        elif ch.isalnum() or ch == '.':
            current += ch
        else:
            if current:
                tokens.append(current.strip())
                current = ''
            # Skip unrecognized characters (like spaces, punctuation)

    if current:
        tokens.append(current.strip())

    return tokens

def extract_math_expressions(tokens):
    expressions = []
    current_expr = []

    for token in tokens:
        if is_operand(token) or is_operator(token):
            current_expr.append(token)
        else:
            if current_expr:
                expressions.append(''.join(current_expr))
                current_expr = []

    if current_expr:
        expressions.append(''.join(current_expr))

    return expressions


```

```python
import nltk
from nltk.corpus import wordnet

nltk.download('wordnet')
#nltk.download('omw-1.4')

def get_synonyms_antonyms(word):
  synonyms = set()
  antonyms = set()

  for syn in wordnet.synsets(word):
    for lemma in syn.lemmas():
      synonyms.add(lemma.name())
      if lemma.antonyms():
        for ant in lemma.antonyms():
          antonyms.add(ant.name())

  return synonyms, antonyms

if __name__ == "__main__":

  word = "sofa"

  synonyms, antonyms = get_synonyms_antonyms(word)

  print(f"Synonyms of '{word}': {synonyms}")
  print(f"Antonyms of '{word}': {antonyms}")
```

```python
import nltk
from nltk.corpus import wordnet
nltk.download('wordnet')
nltk.download('omw-1.4')

def get_hyponyms(word):
  hyponyms = set()

  for syn in wordnet.synsets(word):
    for hypo in syn.hyponyms():
      for lemma in hypo.lemmas():
        hyponyms.add(hypo.name())

  return hyponyms

def get_polysemy(word):
  senses = wordnet.synsets(word)
  return len(senses)

def get_homonyms(word):
  homonyms = set()
  original_senses = wordnet.synsets(word)

  for syn in original_senses:
    for lemma in syn.lemmas():
      if lemma.name() != word:
        other_senses = wordnet.synsets(lemma.name())
        for osyn in other_senses:
          if osyn.name().split('.')[0] == lemma.name() and osyn not in original_senses:
            homonyms.add(osyn.name())

  return homonyms

def display_word_relations(word):
  print(f"\nword: {word}\n{'-'*40}")

  hypos = get_hyponyms(word)
  print(f"Hyponyms: {hypos}")

  senses = wordnet.synsets(word)
  print(f"Polysemy:")
  for i, sense in enumerate(senses, 1):
    print(f"Sense {i}: {sense}")

  homos = get_homonyms(word)
  print(f"Homonyms: {homos}")


if __name__ == "__main__":
  word = input("Enter a word: ").strip().lower()
  display_word_relations(word)
```

## 6-b: Write a function that finds the 50 most frequently occuring words of a text that are not stop words

```python
from collections import Counter
import nltk
from nltk.corpus import stopwords
import re

nltk.download("stopwords")

def top_non_stopwords_nltk(text, n=50):

  stop_words = set(stopwords.words("english"))
  words = re.findall(r'\w+', text.lower())

  filtered = [w for w in words if w not in stop_words]

  return Counter(filtered).most_common(n)

text = "This is not a way to move forward. This way is wrong, and we should not consider it. Moving foward is the best way"

print(top_non_stopwords_nltk(text, n=10))
```

```python
# Draw a bar graph for the wordss and their count

import matplotlib.pyplot as plt
count_words = top_non_stopwords_nltk(text, n=10)

words, counts = zip(*count_words)
plt.bar(words, counts, color='r')
```

```python
# Simple program to compare lemmatization techniques

import nltk
import spacy
from nltk.stem import WordNetLemmatizer
from textblob import Word
import matplotlib.pyplot as plt

nltk.download("punkt")
nltk.download("punkt_tab")
nltk.download("wordnet")

# Increased text to better compare lemmatizers
text = """"The children were running faster than the men.
    She has better shoes but worse luck.
    The geese are flying across the skies.
    "Studies showed that mice were hiding inside the houses."""
tokens = nltk.word_tokenize(text)

# 1. WordNet Lemmatizer
wnl = WordNetLemmatizer()
wordnet_lemmas = [wnl.lemmatize(token) for token in tokens]

# 2. TextBlob Lemmatizer
textblob_lemmas = [Word(token).lemmatize() for token in tokens]

# 3. SpaCy Lemmatizer
nlp = spacy.load("en_core_web_sm")
doc = nlp(text)
spacy_lemmas = [token.lemma_ for token in doc]

# Print results for easier comparison
print("Original Tokens :", tokens)
print("WordNet Lemmas  :", wordnet_lemmas)
print("TextBlob Lemmas :", textblob_lemmas)
print("SpaCy Lemmas    :", spacy_lemmas)

# Calculate the number of words changed by each method
methods = ["WordNet", "TextBlob", "SpaCy"]
changes = [
    sum([t != l for t, l in zip(tokens, wordnet_lemmas)]),
    sum([t != l for t, l in zip(tokens, textblob_lemmas)]),
    sum([t != l for t, l in zip(tokens, spacy_lemmas)])
]

# Draw a bar graph
plt.figure(figsize=(8, 5))
plt.bar(methods, changes, color=["red", "blue", "orange"])
plt.title("Comparison of Lemmatization Methods")
plt.ylabel("Number of Words Changed")
plt.show()
```

```python
import matplotlib.pyplot as plt
from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer

words = ["running", "happiness", "fishing", "easily", "studies", "cries", "nationality"]

porter = PorterStemmer()
lancaster = LancasterStemmer()
snowball = SnowballStemmer("english")

porter_stems = [porter.stem(w) for w in words]
lancaster_stems = [lancaster.stem(w) for w in words]
snowball_stems = [snowball.stem(w) for w in words]

print("Original -> Porter | Lanchester | Snowball\n")
for i, w in enumerate(words):
  print(f"{w:12} -> {porter_stems[i]:8} | {lancaster_stems[i]:9} | {snowball_stems[i]}")

```

```python
fig, ax = plt.subplots(figsize=(10,6))

bar_width = 0.25
x = range(len(words))

ax.bar([i-bar_width for i in x], [len(s) for s in porter_stems], width=bar_width, label="Porter")
ax.bar(x, [len(s) for s in lancaster_stems], width=bar_width, label="Lancaster")
ax.bar([i+bar_width for i in x], [len(s) for s in snowball_stems], width=bar_width, label="Snowball")
```

```python
import nltk
from nltk import pos_tag, word_tokenize, RegexpParser
from nltk.tree import Tree

nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')

text = "The quick brown fox jumps over the lazy dog near the river bank"

tokens = word_tokenize(text)
pos_tags = pos_tag(tokens)

print("Pos Tagged Sentences:\n",pos_tags)
print("-"* 80)

chunk_grammer = r"""
NP: {<DT>?<JJ> * <NN.*>+}  #Optional determiner, adjectives and nouns """

chunk_parser =RegexpParser(chunk_grammer)

chunk_tree = chunk_parser.parse(pos_tags)

print("Chunking Result (Noun Phrases): ")
for subtree in chunk_tree:
  if isinstance(subtree, Tree) and subtree.label() == "NP":
    print(" ".join(word for word, pos in subtree.leaves()))
print("-"*80)

chink_grammer = r"""
    NP: {<.*>+}           # Chunck everthing
        }<VB.*|IN>+{      # Chink out verbs and prepositions
  """

chink_parser = RegexpParser(chink_grammer)
chink_tree = chink_parser.parse(pos_tags)

print("Chinking Result (After Removing Verbs & Prepositions):")
for subtree in chink_tree:
  if isinstance(subtree, Tree) and subtree.label() == "NP":
    print(" ".join(word for word, pos in subtree.leaves()))
print("-"*80)
```
