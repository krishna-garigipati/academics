## Multi Layer Perceptron (MLP)

```python
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.models import Sequential
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
X_train[0].shape
```

```python
model = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])
```

```python
print(model.summary())
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X_train, to_categorical(y_train), validation_split=0.1, epochs=5, batch_size=32)
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
import numpy as np
pred_new = np.argmax(pred, axis=1)
print("Accuracy: ", accuracy_score(y_test, pred_new))
```

```python
print(pred)
```

```python
print(np.argmax(pred, axis=1))
```

```python
loss, accuracy = model.evaluate(X_test, to_categorical(y_test))
```

```python
print(accuracy)
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import matplotlib.pyplot as plt

```

```python
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=10000)
```

```python
def vectorize_sequences(sequences, dimensions=10000):
  results = np.zeros((len(sequences), dimensions))
  for i, sequences in enumerate(sequences):
    results[i, sequences] = 1
  return results
```

```python
X_train = vectorize_sequences(X_train)
X_test = vectorize_sequences(X_test)
```

```python
X_train[0]
```

```python
y_train = np.array(y_train).astype('float32')
y_test = np.array(y_test).astype('float32')
```

```python
model = Sequential([
    Dense(16, activation='relu', input_shape=(10000,)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, validation_split=0.2, epochs=10, batch_size=32)
```

```python
pred = model.predict(X_test)
```

```python
test_loss, test_acc = model.evaluate(X_test, y_test)
```

```python
print(test_acc)
```

```python
plt.plot(history.history['accuracy'], color='b')
plt.title("Accuracy for each epoch")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.show
```

```python
plt.plot(history.history['loss'], color='b')
plt.title("Loss for each epoch")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show
```

```python
hist = model.fit(X_test, y_test, validation_split=0.2, epochs=10, batch_size=32)
```

```python
plt.plot(hist.history['accuracy'], color='b')
plt.title("Accuracy for each epoch: Testing")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.show
```

```python
plt.plot(hist.history['loss'], color='b')
plt.title("Accuracy for each epoch: Testing")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.show
```

```python
from keras.datasets import imdb
from keras.preprocessing import sequence
```

```python
max_features = 10000
maxlen = 500
batch_size = 32
```

```python
(x_train,y_train),(x_test,y_test) = imdb.load_data(num_words = max_features)
```

```python
print(len(x_train),'train sequences')
print(len(x_test),'test sequences')
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
x_train = sequence.pad_sequences(x_train,maxlen = maxlen)
x_test = sequence.pad_sequences(x_test,maxlen = maxlen)
```

```python
from keras.layers import Dense, Embedding, SimpleRNN
from keras.models import Sequential
model = Sequential()
model.add(Embedding(10000,32))
model.add(SimpleRNN(32))
model.add(Dense(1,activation = 'sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer = 'rmsprop',loss = 'binary_crossentropy',metrics = ['accuracy'])
```

```python
history = model.fit(x_train,y_train,epochs = 10,batch_size = 128,validation_split = 0.2)
```

```python
import matplotlib.pyplot as plt
accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1,len(accuracy)+1)
plt.plot(epochs , accuracy , 'bo', label = 'training accuracy')
plt.plot(epochs,val_accuracy,'b' ,label = 'validation accuracy')
plt.title("training and validation accuracy")
plt.legend()
plt.figure()
plt.plot(epochs , loss , 'bo', label = 'training loss')
plt.plot(epochs,val_loss,'b' ,label = 'validation loss')
plt.title("training and validation loss")
plt.legend()
plt.figure()

```

```python
# LSTM
from keras.layers import Dense
from keras.layers import LSTM
model = Sequential()
model.add(Embedding(10000,32))
model.add(LSTM(32))
model.add(Dense(1,activation = 'sigmoid'))
```

```python
model.compile(optimizer = 'rmsprop',loss = 'binary_crossentropy',metrics = ['accuracy'])
```

```python
history = model.fit(x_train,y_train,epochs = 10,batch_size = 128,validation_split = 0.2)
```

```python
import matplotlib.pyplot as plt
accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1,len(accuracy)+1)
plt.plot(epochs , accuracy , 'bo', label = 'training accuracy')
plt.plot(epochs,val_accuracy,'b' ,label = 'validation accuracy')
plt.title("training and validation accuracy")
plt.legend()
plt.figure()
plt.plot(epochs , loss , 'bo', label = 'training loss')
plt.plot(epochs,val_loss,'b' ,label = 'validation loss')
plt.title("training and validation loss")
plt.legend()
plt.figure()
```
